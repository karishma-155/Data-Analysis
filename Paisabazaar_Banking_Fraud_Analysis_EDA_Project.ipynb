## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
def reverse_age(age):
    try:
        # Convert to integer safely (remove decimals)
        age = int(float(age))

        # Reverse only if age < 18
        if age < 18:
            return int(str(age)[::-1])
        return age

    except:
        return age  # return as-is if conversion fails
df["Age"] = df["Age"].apply(reverse_age)


print(df)

````

````python
#Removing NM and replace wtih Not mentioned
df['Payment_of_Min_Amount'] = df['Payment_of_Min_Amount'].replace({'NM': "Not Mentioned"})       #replace NM to not mentioned
df.loc[:,'Payment_of_Min_Amount'].value_counts().reset_index()
````

````python
#Due to multiple entry of person filtring the data so that get the clean for practical benefit
#reduces random fluctuations, and gives a stable, representative value for analysis and modeling.
df= (df.sort_values(['Customer_ID', 'Month']).groupby('Customer_ID').agg({'Name': 'first','Age': 'first','Occupation': 'first','Annual_Income': 'first',
          'Monthly_Inhand_Salary' : 'first','Num_Bank_Accounts' : 'first',
          'Num_Credit_Card' : 'first','Interest_Rate' : 'first',
          'Num_of_Loan' : 'first','Type_of_Loan' : 'first',
          'Delay_from_due_date' : 'mean','Num_of_Delayed_Payment' : 'mean',
          'Changed_Credit_Limit' : 'mean','Num_Credit_Inquiries' : 'first', 'Credit_Mix':'first',
          'Outstanding_Debt' : 'mean','Credit_Utilization_Ratio': 'mean',
          'Credit_History_Age': 'max',
          'Payment_of_Min_Amount' : 'first','Total_EMI_per_month' :'first','Amount_invested_monthly':'first',
          'Monthly_Balance': 'mean','Credit_Score': 'last'}).reset_index())
print(df)
````

````python
# Rounding Numbers into two decimal which making analysis and model results cleaner and more stable.
df['Monthly_Inhand_Salary'] =df['Monthly_Inhand_Salary'].round(2)
df['Annual_Income'] =df['Annual_Income'].round(2)
df['Total_EMI_per_month'] =df['Total_EMI_per_month'].round(2)
df['Amount_invested_monthly'] =df['Amount_invested_monthly'].round(2)
df['Monthly_Balance']=df['Monthly_Balance'].round(2)
df.loc[:,['Monthly_Inhand_Salary', 'Annual_Income', 'Total_EMI_per_month', 'Amount_invested_monthly', 'Monthly_Balance']]
````
